# Python Basics

1. What is Python, and why is it popular

Answer- Python is a high-level, interpreted programming language notable for its clarity and ease of use. It is widely adopted due to its readable syntax, comprehensive standard library, strong community support, cross-platform compatibility, and extensive applications in fields such as web development, data analysis, automation, scripting, and education.

2. What is an interpreter in Python

Answer- An interpreter in Python processes code line by line, translating instructions into machine actions at runtime, thereby eliminating the need for a separate compilation step.

3. What are pre-defined keywords in Python

Answer- Python’s pre-defined keywords—such as if, for, def, class, and import—are reserved for specific syntactic purposes. These keywords cannot be repurposed as variable, function, or class names, as they hold fundamental meaning within the language.

4. Can keywords be used as variable names

Answer- No. Python keywords are reserved and cannot be used as variable/function/class names.

5. What is mutability in Python

Answer- Mutability in Python refers to whether an object’s value can be altered after its creation. Mutable objects, such as lists and dictionaries, allow modification of their contents, whereas immutable types—such as integers, floats, tuples, and strings—do not.

6. Why are lists mutable, but tuples are immutable

Answer- The distinction between lists and tuples lies in their mutability. Lists allow changes to their elements and structure, supporting assignments, insertions, and deletions. Conversely, tuples are immutable and maintain a fixed structure, which enables them to be hashable and suitable for use as dictionary keys.

7. What is the difference between “==” and “is” operators in Python

Answer- The == operator in Python checks whether two objects have equal values, whereas the is operator determines whether two variables reference the same object in memory.

8. What are logical operators in Python

Answer- Logical operators in Python—namely and, or, and not—are used to combine or modify Boolean expressions.

9. What is type casting in Python

Answer- Type casting involves converting a value from one data type to another, such as using int('5'), float(3), or str(10). Implicit type casting occurs automatically when Python evaluates mixed-type expressions (for example, 1 + 2.0 results in 3.0), while explicit type casting requires the programmer to invoke a conversion function directly.

10. What is the difference between implicit and explicit type casting

Answer- Implicit: automatic conversion by Python (e.g., 1 + 2.0 → 3.0).

Explicit: programmer uses conversion functions (int(), float(), str()) to convert.

11. What is the purpose of conditional statements in Python

Answer- Conditional statements, including if, elif, and else, enable selective execution of code based on evaluated conditions.

12. How does the elif statement work

Answer- The elif statement specifically allows for additional condition checks following an initial if statement; the first condition that evaluates as true executes its associated code and subsequent branches are skipped.

13. What is the difference between for and while loops

Answer- For and while loops both facilitate repetition, but differ in usage. For loops iterate over sequences or ranges with a known number of iterations, while while loops continue execution as long as a specified condition remains true, making them more suitable for situations where the number of iterations is undetermined, such as reading input until a certain condition is met.

14. Describe a scenario where a while loop is more suitable than a for loop

Answer- When you don’t know iteration count ahead of time and loop depends on a runtime condition (e.g., reading until end-of-file, repeated input until valid).
